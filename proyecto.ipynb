## Modelo LOG con PCA

### PCA

```python
from sklearn.decomposition import PCA
n_com = 6
pca = PCA(n_components=n_com)
pca.fit(X_train_scaled)
X_pca = pca.transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
X_test_pca_true = pca.transform(X_test_scaled_true)


print(f"Original shape: {X_train_scaled.shape}; Transformed shape: {X_pca.shape}")
```

```python
eigenvectors = pca.components_

components = pca.components_ #eigenvectors (n_componentes, original_dim)
explained_variance_ = pca.explained_variance_ #eigenvalues (n_componentes,)
explained_variance_ratio_ = pca.explained_variance_ratio_

columns_pca = []
var_num = 0
for i,n in enumerate(explained_variance_ratio_):
    print(f"La varianza en el componente principal {i+1} es de {explained_variance_ratio_[i]*100:.2f} %")
    columns_pca.append(f"Componente principal {i+1}")
    var_num += explained_variance_ratio_[i]*100
# Display the eigenvectors in the console
print("Eigenvectors (principal components):")
for i, eigenvector in enumerate(components):
    print(f"Principal Component {i+1}: {eigenvector}")

```

```python
print(f"La varianza comprendida es de {var_num:.2f} %")
```
